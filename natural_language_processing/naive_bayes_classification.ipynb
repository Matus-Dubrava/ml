```python
import nltk
import random
from nltk.corpus import movie_reviews, stopwords
```

```python
documents = []

for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((list(movie_reviews.words(fileid)), category))
```

```python
documents[0]
```

```python
all_words = []
for w in movie_reviews.words():
    all_words.append(w.lower())
    
all_words = nltk.FreqDist(all_words)
```

```python
print(len(all_words))
print(all_words.most_common(15))
print(all_words['stupid'])
```

```python
# picking top 3000 most frequent words
word_features = [x[0] for x in all_words.most_common(3000)]
word_features[:10]
```

```python
def find_features(document):
    words = set(document)
    features = {}
    
    for w in word_features:
        features[w] = (w in words)
        
    return features
```

```python
print(list(find_features(movie_reviews.words('neg/cv000_29416.txt')).items())[:5])
```

```python
featureset = [(find_features(rev), category) for rev, category in documents]
featureset[:5]
```

```python
training_set = featureset[:1900]
testing_set = featureset[1900:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
print(f'Naive Bayes Algo accuracy: {nltk.classify.accuracy(classifier, testing_set)}')
classifier.show_most_informative_features(15)
```
