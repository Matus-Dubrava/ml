# interaction term

```r
summary(lm(medv~lstat*age, data=Boston)) # shorthand for medv~lstat+age+lstat*age
```

```r
summary(lm(medv~lstat:age, data=Boston)) # include only the interaction term
```

# non-linear transfromation of the predictors

```r
lm.fit2 = lm(medv~lstat+I(lstat^2), data=Boston)
summary(lm.fit2)
```

```r
plot(Boston$lstat, Boston$medv)
```

# model comparison (anova)

```r
lm.fit = lm(medv~lstat, data=Boston)
anova(lm.fit, lm.fit2)
```

```r
par(mfrow=c(2, 2))
plot(lm.fit2)
```

```r
lm.fit5 = lm(medv~poly(lstat, 5), data=Boston)
summary(lm.fit5)
```

```r
summary(lm(medv~log(rm), data=Boston))
```
