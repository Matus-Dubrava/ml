```r
# how many suburbs with average number of rooms per dwelling greater than 7 and 8
rm_gt_7 = rep(0, nrow(Boston))
rm_gt_7[Boston$rm > 7] = 1
Boston$rm_gt_7 = rm_gt_7

rm_gt_8 = rep(0, nrow(Boston))
rm_gt_8[Boston$rm > 8] = 1
Boston$rm_gt_8 = rm_gt_8
```

```r
sum(Boston$rm_gt_7)
```

```r
sum(Boston$rm_gt_8)
```

```r
# find suburbs with lowest median value of owner occupied homes
Boston[Boston$medv == min(Boston$medv), ]
```

```r
apply(Boston, 2, range)
```

```r
# How many suburbs in this data set bound the Charles River?
nrow(Boston[Boston$chas == 1, ]) 
```

```r
# suburbs with high crime ratio per capita
BostonHighCrime = Boston[Boston$crim >= quantile(Boston$crim, 0.95), ]
```

```r
nrow(BostonHighCrime) / nrow(Boston)
```
