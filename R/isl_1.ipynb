```r
x = seq(0, 1, length=100) # 100 equally distributed points between 0 and 1
x
```

```r
y = x
f = outer(x, y, function(x, y) cos(y) / (1 + x^2))
contour(x, y, f)
contour(x, y, f, nlevels=45, add=T)
fa = (f - t(f)) / 2
contour(x, y, fa, nlevels=15)
```

```r
image(x, y, fa) # heat map
```

```r
persp(x, y, fa, theta=30, phi=30) # 3D plot, theta and phi controls view angle
```
