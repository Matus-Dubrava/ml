# Covariance & Correlation

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")
```

```python
def de_mean(x):
    mean = np.mean(x)
    return [xi - mean for xi in x]

def covariance(x, y):
    n = len(x) 
    return np.dot(de_mean(x), de_mean(y)) / (n - 1)
```

```python
page_speeds = np.random.normal(3.0, 1.0, 1000)
purchase_amount = np.random.normal(50.0, 10.0, 1000)
plt.scatter(page_speeds, purchase_amount, edgecolor="k");
```

```python
covariance(page_speeds, purchase_amount)
```

```python
purchase_amount = np.random.normal(50.0, 10.0, 1000) / page_speeds

plt.scatter(page_speeds, purchase_amount, edgecolor="k");
```

```python
covariance(page_speeds, purchase_amount)
```

```python
def correlation(x, y):
    x_std = np.std(x)
    y_std = np.std(y)
    return covariance(x, y) / x_std / y_std # possible division by zero!!!
```

```python
correlation(page_speeds, purchase_amount)
```

```python
np.corrcoef(page_speeds, purchase_amount)
```
