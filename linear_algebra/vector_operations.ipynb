# VECTORS

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
style.use('ggplot')
```

```python
# vector addition
# x = [x1, x2]
# y = [y1, y2]
# v1 + v2 = [x1 + y1, x2 + y2]

x = np.array([-1, 2])
y = np.array([4, 2])
r = x + y
print(r)

plt.plot([0, x[0]], [0, x[1]], color='green', label='x')
plt.plot([0, y[0]], [0, y[1]], color='blue', label='y')
plt.plot([0, r[0]], [0, r[1]], color='red', label='x + y')
plt.legend()
plt.show()
```

```python
# vector subtraction
# x = [x1, x2]
# y = [y1, y2]
# v1 - v2 = [x1 - y1, x2 - y2]

x = np.array([4, 2])
y = np.array([-1, 2])
r = x - y
print(r)

plt.plot([0, x[0]], [0, x[1]], color='green', label='x')
plt.plot([0, y[0]], [0, y[1]], color='blue', label='y')
plt.plot([0, -y[0]], [0, -y[1]], color='blue', label='-y', linestyle='--')
plt.plot([0, r[0]], [0, r[1]], color='red', label='x + y')
plt.legend()
plt.show()
```

```python
# vector multiplication

x = np.array([4, 2])
c = 2
r = x * 2

print(r)

plt.plot([0, x[0]], [0, x[1]], color='green', label='x')
plt.plot([0, r[0]], [0, r[1]], color='red', label='x * c', linestyle='--')
plt.legend()
plt.show()
```
