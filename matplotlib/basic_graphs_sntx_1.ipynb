# SIMPLE LINE GRAPH


#### legend
Specify label in plt.plot and call plt.legend()

```python
# %matplotlib notebook
import matplotlib.pyplot as plt

x = [1,2,3]
y = [5,7,4]

x2 = [1,2,3]
y2 = [10,14,12]

plt.plot(x, y, label='First Line')
plt.plot(x2, y2, label='Second Line')
plt.title('My Title\nCheck it out')
plt.xlabel('Plot Number')
plt.ylabel('Importanat Var')
plt.legend()
plt.show()
```

# BAR CHART

```python
x = [0,2,4,6,8]
y = [6,7,8,2,4]

x2 = [1,3,5,7,9]
y2 = [4,2,4,1,5]

plt.bar(x, y, label='bars1', color='dodgerblue')
plt.bar(x2, y2, label='bars2', color='salmon')
plt.legend()
plt.show()
```

# HISTOGRAM

# SCATTER PLOT

```python
x = [1,2,3,4,5,6,7,8]
y = [5,4,2,1,3,5,2,1]

plt.scatter(x, y, label='scatter1', color='salmon', marker='x', s=150)
plt.legend() 
```

# STACK PLOT

```python
days =     [1,2,3,4,5]
sleeping = [7,8,6,11,7]
working =  [7,8,7,2,2]
eating =   [2,3,4,3,2]
playing =  [8,5,7,8,13]

plt.plot([], [], color='salmon', label='sleeping', linewidth=5)
plt.plot([], [], color='cyan', label='eating', linewidth=5)
plt.plot([], [], color='red', label='working', linewidth=5)
plt.plot([], [], color='black', label='playing', linewidth=5)

plt.stackplot(days, sleeping, eating, working, playing, colors=['salmon', 'cyan', 'red', 'black'])
plt.title('My Title\nCheck it out')
plt.legend()
plt.show()
```

# PIE CHART

```python
days =     [1,2,3,4,5]
sleeping = [7,8,6,11,7]
working =  [7,8,7,2,2]
eating =   [2,3,4,3,2]
playing =  [8,5,7,8,13]

slices = [7,2,2,13]
activities = ['sleeping', 'eating', 'working', 'playing']
colors = ['salmon', 'cyan', 'red', 'dodgerblue']

plt.pie(slices, 
        labels=activities, 
        colors=colors, 
        startangle=90, 
        shadow=True, 
        explode=(0,0.1,0,0),
        autopct='%1.1f%%')

plt.show()
```
