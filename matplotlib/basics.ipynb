```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import style
style.use('ggplot')
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y, color='skyblue')
plt.show()
```

```python
# 1st method
fig = plt.figure()
ax = fig.add_subplot() 
plt.show()
```

```python
# 2nd method
fig = plt.figure() 
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd method
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, [50, 100, 200, 250], color='skyblue') 
ax.set(title='simple plot', xlabel='x-axis', ylabel='y-axis')
plt.show()

fig.savefig('../datasets/images/sample-plot.png')
```

## figures with np arrays

```python
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(12, 12), nrows=2, ncols=2)

# line plot
ax[0][0].plot(x, x**2, color='dodgerblue')
ax[0][0].set(title='line graph')

ax[0][1].scatter(x, np.exp(x), color='dodgerblue')
ax[0][1].set(title='scatter plot (exp)')

ax[1][0].scatter(x, np.sin(x), color='dodgerblue')
ax[1][0].set(title='scatter plot (sin)')
plt.show()
```

```python
# make a plot from dictionary
nut_butter_prices = {
    'almond butter': 10,
    'peanut butter': 8,
    'cashew butter': 12
}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values(), color='dodgerblue')
ax.set(title='Nut Butter', ylabel='Price')
plt.show()
```

```python
# horizontal bar plot
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()), color='dodgerblue')
plt.show()
```

```python
# historgrams
x = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(x, color='dodgerblue')
plt.show()
```

```python
# subplots 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
ax1.plot(x, x/2, color='dodgerblue')
ax2.scatter(np.random.random(10), np.random.random(10), color='dodgerblue')
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values(), color='dodgerblue')
ax4.hist(np.random.randn(1000), color='dodgerblue')
plt.show()
```

```python
heart_disease = pd.read_csv('../datasets/heart-disease.csv')
heart_disease.head(3)
```

```python
# plot ditribution of age column
fig, ax = plt.subplots()
ax.hist(heart_disease['age'], color='dodgerblue', bins=30)
plt.show()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50['age'], 
                     y=over_50['chol'], 
                     c=over_50['target'], 
                     cmap='winter', 
                     edgecolor='k')

ax.axhline(y=over_50['chol'].mean(), linestyle='--', label='mean cholesterol')

ax.set(title='Heart Disease and Cholersterol', 
       xlabel='age', 
       ylabel='cholesterol')

ax.legend(*scatter.legend_elements(), title='Target')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, figsize=(12, 12), sharex=True)

scatter_1 = ax[0].scatter(x=over_50['age'], 
                          y=over_50['chol'], 
                          c=over_50['target'],
                          cmap='winter',
                          edgecolor='k')

ax[0].axhline(y=over_50['chol'].mean(), linestyle='--')
ax[0].set(title='Heart Disease & Cholesterol', ylabel='Cholesterol')
ax[0].legend(*scatter_1.legend_elements(), title='target')

scatter_2 = ax[1].scatter(x=over_50['age'], 
                          y=over_50['thalach'], 
                          c=over_50['target'], 
                          cmap='winter', 
                          edgecolor='k')

ax[1].axhline(y=over_50['thalach'].mean(), linestyle='--')
ax[1].set(title='Heart Disease & Thalach', xlabel='Age', ylabel='Thalach')
ax[1].legend(*scatter_1.legend_elements(), title='target')

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
plt.show()
```
